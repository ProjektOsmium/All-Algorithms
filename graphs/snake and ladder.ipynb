```python
from collections import defaultdict
```

```python
adjList = defaultdict(lambda:[])
for i in range(0,100):
    if i+1<101:
        adjList[i].append(i+1)
    if i+2<101:
        adjList[i].append(i+2)
    if i+3<101:
        adjList[i].append(i+3)
    if i+4<101:
        adjList[i].append(i+4)
    if i+5<101:
        adjList[i].append(i+5)
    if i+6<101:
        adjList[i].append(i+6)
```

```python
def enter_ladder_snake(start,end):
    if start<=99 and end<=100:
        if end not in adjList[start]:
            adjList[start]=[end]        # replace normal dice moves by ladder or snake move
```

```python
path = defaultdict(lambda:float("inf"))
def bfs(adj,source,destination):
    path[source] = 0
    visited = defaultdict(lambda:False)
    q = [source]
    while len(q)!=0:
        node = q.pop(0)
        for nextnode in adj[node]:
            if not visited[nextnode]:
                if len(adj[node])==1:            # check if node is a ladder or snake
                    path[nextnode] = path[node]  
                else:
                    path[nextnode] = path[node]+1
                q.append(nextnode)
                visited[nextnode]=True
    return path[destination]
```

```python
# 1st roll: 0-6
# 2nd roll: 6-(12-50) (move up the ladder from 12 to 50)
# 3rd roll: 50-(54-88) (move up the ladder from 54 to 88)
# 4th roll: 88-93
# 5th roll: 93-99
# 6th roll: 99-100
bfs(adjList,0,100)
```
